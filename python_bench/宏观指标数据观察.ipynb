```python
import akshare as ak
import pandas as pd
import matplotlib
```

### 1. PMI

#### 1.1 获取PMI数据: pmi_df

```python
pmi_df = ak.macro_china_pmi_yearly()
pmi_df = pmi_df[['日期', '今值']]
pmi_df.columns = ['date', 'pmi']
pmi_df.reset_index()
pmi_df['ma3'] = pmi_df['pmi'].rolling(window=3).mean()
pmi_df = pmi_df.set_index('date')
pmi_df.plot(kind='line', figsize = (25, 15))
```

### 2. 上海银行间同业拆放利率（Shibor_1M）

```python
shibom_1M_df = ak.macro_china_shibor_all()
shibom_1M_df = shibom_1M_df[['日期', '1M-定价']]
shibom_1M_df.columns = ['date', '1M-Price']
shibom_1M_df = shibom_1M_df.set_index('date')
shibom_1M_df.plot(kind='line', figsize = (25, 15))
```
